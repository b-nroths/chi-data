```python
import pyarrow.parquet as pq
import s3fs, json
S3FS = s3fs.S3FileSystem()
ds = pq.ParquetDataset(
		path_or_paths='bnroths/chicago-data/%s/year=2001' % 'crimes', 
		filesystem=S3FS, 
		validate_schema=False
		)
```

```python
table = ds.read()
```

```python
df = table.to_pandas()
```

```python
df['dt'] = df['date'].str[:7]
```

```python
groups = dict(list(df.groupby('dt')))
```
